```python
def water_jug_dfs(a, b, target):
    visited = set()

    def dfs(x, y):
        if (x, y) in visited:
            return False
        print(f"Current State: ({x}, {y})")
        if x == target:
            print("Target Achieved!")
            return True
        visited.add((x, y))
        # Rule 1: Fill Jug A
        if dfs(a, y):
            print(f"Rule Applied: Fill A -> ({a}, {y})")
            return True
        # Rule 2: Fill Jug B
        if dfs(x, b):
            print(f"Rule Applied: Fill B -> ({x}, {b})")
            return True
        # Rule 3: Empty Jug A
        if dfs(0, y):
            print(f"Rule Applied: Empty A -> (0, {y})")
            return True
        # Rule 4: Empty Jug B
        if dfs(x, 0):
            print(f"Rule Applied: Empty B -> ({x}, 0)")
            return True
        # Rule 5: Pour A to B
        t = min(x, b - y)
        if dfs(x - t, y + t):
            print(f"Rule Applied: Pour A→B -> ({x - t}, {y + t})")
            return True        
        # Rule 6: Pour B to A
        t = min(y, a - x)
        if dfs(x + t, y - t):
            print(f"Rule Applied: Pour B→A -> ({x + t}, {y - t})")
            return True

        return False
    
    if not dfs(0, 0):
        print("No Solution Found")


a = int(input("Capacity A: "))
b = int(input("Capacity B: "))
target = int(input("Target in A: "))

water_jug_dfs(a, b, target)
```
