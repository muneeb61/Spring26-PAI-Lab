#### ___N Queen Problem___

- ___Problem of Chess (Dynamically)___

```python
def c_borad(n):
    board = [-1] * n
    solutions = [0]
    
    def is_safe(row, col):
        for i in range(row):
            if board[i] == col or abs(board[i] - col) == abs(i - row):
                return False
        return True
    
    def print_solution():
        print(f"\nSolution {solutions[0]}:")
        for i in range(n):
            row = ""
            for j in range(n):
                if board[i] == j:
                    row += "Q "
                else:
                    row += ". "
            print(row)
    
    def solve(row):
        if row == n:
            solutions[0] += 1
            print_solution()
            return
        for col in range(n):
            if is_safe(row, col):
                board[row] = col
                solve(row + 1)
                board[row] = -1
    
    solve(0)
    print(f"\nTotal solutions for {n}-Queens: {solutions[0]}")

n = int(input("Enter value of N: "))
c_borad(n)
```
